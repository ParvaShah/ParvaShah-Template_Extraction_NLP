```
import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_sm")
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
```

```
sentence = "Steven Paul Jobs \
was an American business magnate and investor. He was the \
chairman, chief executive officer (CEO), and co-founder of Apple Inc.; \
chairman and majority shareholder of Pixar; a member of The Walt \
Disney Company's board of directors following its acquisition of \
Pixar; and the founder, chairman, and CEO of NeXT. \
"
doc_sent = nlp(sentence)
```

```
print('Input Sentence: ',doc_sent)
print()
for i,sent in enumerate(doc_sent.sents):
  print(i,'Sentence: ',sent)
  doc_token = nlp(str(sent))
  tokens = [tokens for tokens in doc_token]
  print('Tokens: ',tokens)
  lemma = [tokens.lemma_ for tokens in doc_token ]
  print('Lemmas: ',lemma)
  pos = [tokens.pos_ for tokens in doc_token ]
  print('POS: ',pos)
  dep = [tokens.dep_ for tokens in doc_token ]
  print('DEP: ',dep)
  hypernyms_list = []
  for token in tokens:
    tmp = []
    for ss in wn.synsets(token.text):
        if ss.hypernyms():
            tmp.append(ss.hypernyms()[0].name()[:-5])
    hypernyms_list.append(tmp)
  print('Hypernymy: ',hypernyms_list)
  
  hyponyms_list = []
  for token in tokens:
    tmp = []
    for ss in wn.synsets(token.text):
        if ss.hyponyms():
            tmp.append(ss.hyponyms()[0].name()[:-5])
    hyponyms_list.append(tmp)
  print('Hyponyms: ',hyponyms_list)

  meronyms_list = []
  for token in tokens:
    tmp = []
    for ss in wn.synsets(token.text):
        if ss.part_meronyms():
            tmp.append(ss.part_meronyms()[0].name()[:-5])
    meronyms_list.append(tmp)
  print('Meronyms: ',meronyms_list)
  
  holonyms_list = []
  for token in tokens:
    tmp = []
    for ss in wn.synsets(token.text):
        if ss.part_holonyms():
            tmp.append(ss.part_holonyms()[0].name()[:-5])
    holonyms_list.append(tmp)
  print('Holonyms: ',holonyms_list)
  
  print()

```

```
sentence = "In 2017, Amazon acquired Whole Foods Market for \
US$13.4 billion, which vastly increased Amazon's presence as a brickand-mortar retailer."
displacy.render(nlp(sentence), style='dep', jupyter=True, options={'distance': 90})
```
